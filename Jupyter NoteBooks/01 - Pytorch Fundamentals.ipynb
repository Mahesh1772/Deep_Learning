## Stacking, squeezing and unsqueesing

View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor

Squeeze - removes all `1` dimensions from a tensor
Unsqueeze - add a `1` dimension to a target tensor
Permute - Return a view of the input with dimensions permuted (swapped)

```python
x = torch.arange(1,10)
x, x.shape
```

```python
x.reshape(1,9) # 2D array # Must be compatable with the the total number of elements
```

```python
z = x.view(1,9) # x and z share the same memory space, change in one changes both, z added another bracket. now is a 2d vector
z
```

```python
z[:, 0] = 5
z, x
```

```python
torch.stack([x, x, x], dim= 0)
```

```python
torch.stack([x, x, x], dim=1)
```

```python
torch.unsqueeze(x, dim=0)
```

```python
# squeese - Removed on square bracket, all the single dimensions
z, z.squeeze(), z.shape, z.squeeze().shape
```

##If the dimensions are : 1, 3, 1
##After squeeze: 3 - It will remove every dimension where the value of it is 1

#Unsqueeze

### Dim= 0 adds 1 dimensions before existing dimension {1 in the front}
### Dim=1 adds 1 dimension after the existing

```python
z = z.squeeze()

print(f"Old tensor: {z}")
print(f"Old shape: {z.shape}")
print()

print(f"New tensor: {z.unsqueeze(dim=0)}")
print(f"New shape: {z.unsqueeze(dim=0).shape}")
print()

print(f"Newer tensor: {z.unsqueeze(dim=1)}")
print(f"Newer shape: {z.unsqueeze(dim=1).shape}")
```

#Permute

## The permuted image and the image share same space. Same as view

```python
image = torch.rand(size=(224,224,3)) # height, width, color

permuted = image.permute(2, 0, 1) # Shift the 2nd dimension(3) to the first position (224)
permuted.shape # color, height, width
```

```python
image[0, 0, 0] = 2323
image [0, 0, 0], permuted[0, 0, 0]
```
