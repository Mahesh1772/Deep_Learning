We'll use a linear regression formula to make a straight line with *known* **parameters**

```python
# Creating known parameters
weight = 0.7
bias = 0.3

# create
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1) # adds 1 dimension to the column making it a col vector
y = weight * X + bias

X[:10], y[:10]
```

```python
train = int(0.8 * len(X))
X_train, y_train, = X[:train], y[:train] # first 80 percent is train data
X_test, y_test = X[train:], y[train:] # last 20 percent is test data

len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
def plot_predictions(train_data = X_train,
                     train_labels= y_train,
                     test_data = X_test,
                     test_labels = y_test,
                     predictions = None):
  plt.figure(figsize=(10, 7))

  # Plot training data in blue
  plt.scatter(train_data, train_labels, c='b', s=4, label = "Training data")

  # Plot test data in green
  plt.scatter(test_data, test_labels, c='r', s=4, label= "test data")

  if predictions is not None:
    #plot predictions
    plt.scatter(test_data, predictions, c='g', s=4, label = 'Predictions')

  plt.legend(prop={"size": 14})
```

## Building a model

```python
# Linear regression model
from torch import nn
```

```python
##
class LinearRegression(nn.Module): # <- almost everthing in nueral network will need to subclass this class
  def __init__(self):
    super().__init__()
    # gradient descent = true helps nn to change the weights and bias progressively as we go if and when needed using: Backpropagation and gradient descent

    # Initialize model parameters - Can be single value as seen in the code OR can be multiple layer ones
    self.weights = nn.Parameter(torch.rand(1,
                                           requires_grad=True, # <- Uses torch.auto grad for gradient descent calculations
                                           dtype=torch.float)) #Random weights
    self.bias = nn.Parameter(torch.rand(1,
                                        requires_grad=True,
                                        dtype=torch.float)) # Random bias

    #Forward method to define the computation in the model: NEEDS TO BE OVERRIDEN
  def forward(self, x: torch.Tensor) -> torch.Tensor: # <-"x" is the input data
      return self.weights * x + self.bias
```

### Pytorch model building essentials

torch.nn - contails building block

torch.nn.Module - Base class for all nueral networks modules, if subclassed, must override forward()

torch.nn.Parameter - Parameters learned by the model, set by Pytorch layer from torch.nn

torch.optim - optimiser which help with grad descent
def forward() - Used for computation of model

## Training model

Moving model fron unkown parameter (bad predictions) to known parameters (good predictions)

Loss function - Measures the difference between predicted and actual output

Optimizer - Uses loss fucntion and adjust model parameters, so that the loss functions is minimized.

We need:
* training looops
* test loops

```python
# Loss function
loss_fn = nn.L1Loss()

# Setup optimizer
optimizer = torch.optim.SGD(params=model.parameters(), lr= 0.001)
```

### Training loop and Testing looop

Things include:
* Loop through the data
* forward pass (data moving through the forward functions, or from input through hidder layer to output layer)
* Calculate the loss (compare forward pass predictions to truth )
* Optimiser zero grad
* loss backward (data moves backwards through the network to calculate the gradients of each of the parameter w.r.t the loss) (**back propagation**)
* Optimizer step - use the optimiser to adjust our model's parameter (**gradient descent**)

```python
# Epochs - loop through data : Hyper parameters
torch.manual_seed(42)

epochs = 1000

# Used for tracking model progress throughout epochs
epoch_count = []
loss_values = []
test_loss_values = []

for epoch in range(epochs):
  #0. set the model to training mode, default mode
  model.train() # Makes sure the gradient descent calculation is turned on

  #1. forward pass, gather predictions
  y_pred = model(X_train)

  #2. Calculating loss b/w test and predictions data
  loss = loss_fn(y_pred, y_train)

  #3. Optimizer 0 grad as they accumulate by default, BUT we only want to caluculate for each iteration
  optimizer.zero_grad()

  #4. Back propagation - calculate the gradients of each parameter
  loss.backward()

  #5. step the optimizer (perform gradient descent)
  optimizer.step()

  ###Testing = turn off settings not needed for testing (dropout layer, batchnorm layers)
  model.eval()
  with torch.inference_mode(): #off the gradient tracking (step 4 and 5 will not cause changes)

    #1. Forward pass
    test_pred = model(X_test)

    #2. Calculate test loss
    test_loss = loss_fn(test_pred, y_test)

  if epoch % 10 == 0:
    epoch_count.append(epoch)
    loss_values.append(loss)
    test_loss_values.append(test_loss)
    print(f"Epoch: {epoch} | Test Loss {test_loss} | Loss: {loss}")
    model.state_dict()
```

#EVERYTHING ALL IN ONE

```python
import torch
from torch import nn
import matplotlib.pyplot as plt

# Device agnostic code
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device

# Create some data using linear regression formula: y = bias + weight * X
weight = 0.5
bias = 0.8

# Create range values
start = 0
end = 1
step = 0.02

# Create X and y
X = torch.arange(start, end, step).unsqueeze(dim=1)
y = bias + weight * X

# Split data
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]

def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    plt.figure(figsize=(10, 7))

    # Plot training data in blue
    plt.scatter(train_data, train_labels, c='b', s=4, label="Training data")

    # Plot test data in red
    plt.scatter(test_data, test_labels, c='r', s=4, label="Test data")

    if predictions is not None:
        # Plot predictions in green
        plt.scatter(test_data, predictions, c='g', s=4, label='Predictions')

    plt.legend(prop={"size": 14})
    plt.show()

plot_predictions(X_train, y_train, X_test, y_test)

class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layer = nn.Linear(in_features=1, out_features=1)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)

# Set manual seed
torch.manual_seed(42)
model_1 = LinearRegressionModel()
model_1.to(device)

# Loss function
loss_fn = nn.L1Loss()  # Same as MAE

# Optimizer
optimizer = torch.optim.SGD(lr=0.001, params=model_1.parameters())

# Training loop
epochs = 2000

# Device agnostic code for DATA
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

for epoch in range(epochs):
    model_1.train()

    # 1. Predictions
    y_pred = model_1(X_train)

    # 2. Loss calculation
    loss = loss_fn(y_pred, y_train)

    # 3. Gradient set to zero
    optimizer.zero_grad()

    # 4. Backpropagation
    loss.backward()

    # 5. Optimizer step
    optimizer.step()

    ###################### Testing ################################
    model_1.eval()

    with torch.inference_mode():
        # 1. Forward pass
        y_test_pred = model_1(X_test)

        # 2. Test loss
        test_loss = loss_fn(y_test_pred, y_test)

    if epoch % 100 == 0:
        print(f"Epoch: {epoch} | Loss: {loss} | Test Loss: {test_loss}")

# Move predictions back to CPU for plotting
y_test_pred = y_test_pred.cpu()

# Plot predictions
plot_predictions(predictions=y_test_pred)
```
