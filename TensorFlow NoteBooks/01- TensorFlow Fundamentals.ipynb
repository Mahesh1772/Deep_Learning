## Introduction to Tensor

- Scalar: 0 dim
- Vecotr: 1 dim
- Matrix: 2 dim
- Tensor: <= 3 dim

```python
# Import TensorFlow
import tensorflow as tf
print(tf.__version__)
```

```python
# Create tensors with tf.constant()
scalar = tf.constant(7)
scalar
```

```python
# Check the number of dimensions on a tensor
scalar.ndim
```

```python
# Create a vector
vector = tf.constant([10,10])
vector
```

```python
# Check the dimension of our vector
vector.ndim
```

```python
# Create a matrix
matrix = tf.constant([[10, 8],
                      [10, 7]])
matrix, matrix.ndim
```

```python
# Create another matrix
another_matrix = tf.constant([[10., 4.],
                              [3., 1.],
                              [8., 9.]], dtype=tf.float16)
another_matrix
```

```python
another_matrix.ndim
```

```python
# Tensor
tensor = tf.constant([[[1,2,3], [2,3,4]], [[3,4,5], [3, 4, 5]]])
tensor
```

```python
tensor.ndim
```

### Getiing informaiton form tensors
- Shape: len of each dimemntion
- Rank: num of dimemensions
- Axis: Particular dimention of tensor
- Size: Total number of items in a tensor

```python
# Create a rank 4 tensor
rank_4_tensor = tf.zeros((2,3,4,5)) # last dim -> first dim
```

```python
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
# Get various attributes of our tensor
print('Datatypes of every element:', rank_4_tensor.dtype)
print('Number of dimension(rank): ', rank_4_tensor.ndim)
print("Shape of tensor: ", rank_4_tensor.shape)
print("Elements along the 0 axis: ", rank_4_tensor.shape[0])
print("Elements along the last axis: ", rank_4_tensor.shape[-1])
print('Total numbher of elements in the tensor: ', tf.size(rank_4_tensor).numpy())
```

```python
rank_2_tensor = tf.ones((2, 4))
rank_2_tensor
```

```python
rank_3_tensor = rank_2_tensor[..., tf.newaxis] # For every axis before the last one
rank_3_tensor
```

```python
tf.expand_dims(rank_2_tensor, axis=-1) # Expand on which dimension
```

```python
tf.expand_dims(rank_2_tensor, axis=0) # Expand on which dimension
```

### One-hot encoding tesnors

```python
# Create a list of indicies
some_list = [0,1,2,3,1] # red, green, blue, yellow

tf.one_hot(indices=some_list, depth=len(some_list))
```

```python
# Speify custom values for one hot encoding
tf.one_hot(some_list, len(some_list), on_value='This', off_value='Nah')
```
